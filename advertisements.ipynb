```python
amount_years_to = []
for col in ['Newspaper Total', "Magazine Total", "Television Total", "Total Radio","Direct Mail","Internet"]:
  filtered_df_initial = df2[(df2[col]>=600)]
  initial_val = filtered_df_initial.iloc[0]["Year"]
  filtered_df_final = df2[(df2[col]>= 1050)]
  final_val = filtered_df_final.iloc[0]["Year"]
  amount_years_ = final_val - initial_val
  amount_years_to.append(amount_years_)
  filtered_df_initial = pd.DataFrame
  filtered_df_final = pd.DataFrame
amount_years_to

  
```

```python
data = {"Media":['Newspaper Total', "Magazine Total", "Television Total", "Total Radio","Direct Mail","Internet"],'Amount of Years': amount_years_to}
new_difference_tbl = pd.DataFrame(data)
new_difference_tbl =new_difference_tbl.sort_values("Amount of Years",ascending=False)
```

```python
ax = plt.subplots()
hbars = plt.barh(new_difference_tbl["Media"],new_difference_tbl["Amount of Years"])
plt.xlabel("Amount of Years")
plt.ylabel("Medium")
plt.title("Amount of Years it Took to go from 600 Million to 1050 Million")


plt.show()
```
